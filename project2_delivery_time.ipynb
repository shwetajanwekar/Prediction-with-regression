```python
##importing libraries
import pandas as pd
import seaborn as sns
import time
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score
print("Import the libraries Successfully")
st = time.time()

## reading Data
print("Reading of Data")
def read_data():
    try:
       data1=pd.read_csv("E:/delivery_time.csv")
       print(data1.head(2))
       return data1
    except Exception as e:
       print("Error in read data function_",e)
##call the function
data1=read_data()

## data profiling
def data_profiling(data1):
    try:
        records=len(data1)
        columns=data1.columns
        missing=data1.isnull().sum()
        data_types=data1.dtypes
        stats=data1.describe
        print("number of records-",records)
        print("columns-",columns)
        print("missing-\n",missing)
        print(data_types)
        print(stats)
    except Exception as e:
        print("Error in data profile function-",e)
#calling function
data_profiling(data1)

##Exploratory data Analysis
def exploratory_analysis(data1):
    try:
        plt.figure(figsize=(7,5))
        ax=sns.displot(x=data1['Delivery Time'])
        plt.figure()
        ax1=sns.displot(x=data1['Sorting Time'])
        print("sortingtime_skewness",data1['Sorting Time'].skew())
        print("deliverytime_skewness",data1['Delivery Time'].skew())
        corr=data1.corr()
        plt.figure()
        ax3=sns.heatmap(corr,annot=True)
        plt.show()
        var=data1.var()
        print(var)
    except Exception as e:
        print("Error in exploratory analysis-",e)
#calling function
exploratory_analysis(data1)

##data splitting

def data_splitting(data1):
    try:
        X=data1['Delivery Time'].values.reshape(-1,1)
        y=data1['Sorting Time'].values.reshape(-1,1)

#splitting of data
        X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.15, random_state=1234)
        return X_train,X_test,y_train,y_test
    except Exception as e:
        print("Errors in data splitting-",e)
X_train,X_test,y_train,y_test = data_splitting(data1)  

##data modelling
def data_modelling(X_train, y_train):
    global model
    model = LinearRegression()
    model.fit(X_train, y_train)
    print("training is done-")
    coeff=model.coef_
    constant=model.intercept_
    score=model.score(X_train,y_train)
    print("Score for training model",score)
    return coeff,constant
#calling function
coeff,constant=data_modelling(X_train, y_train)  

##prediction and evaluation on predicted data
def model_evaluation(X_test, y_test):
    predict= model.predict(X_test)
    final_score =r2_score(y_test,predict)
    return predict,final_score
predict,final_score=model_evaluation(X_test, y_test)
print("Prediction-\n",predict)
print("Final score-",final_score)

en=time.time()
print("Total time taken in second -",en-st)


        





```
